```python
# Construisons un seul dataset à partir des 2 premiers pour avoir un peu plus de visibilité

df = list_of_orders.merge(order_details, on = 'Order ID')
df.head(10), df.shape
```

```python
df.isna().sum()
```

```python
# Définition correcte de la date dans le dataset + création d'une variable month et year

df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst = True)

df['Month'] = df['Order Date'].dt.month
df['Year'] = df['Order Date'].dt.year

df.head()
```

```python
# Maintenant regardons en détails les produits que l'on vend par catégorie, le CA et le profit associé

df_cat_sub = (df.groupby(['Category', 'Sub-Category'], as_index = False).agg({
    'Amount':'sum',
    'Quantity': 'sum',
    'Profit': 'sum'   
    })
)
df_cat_sub['Margin'] = df_cat_sub['Profit'] / df_cat_sub['Amount']
df_cat_sub
```

```python
# Insights business
# 1 - La sous-catégorie 'T-shirt' génère le plus de bénéfice
# 2 - Certaines sous-catégorie comme 'Printers', 'Bookcases' ou encore 'Saree' génèrent beaucoup de CA
# 3 - Certaines sous-catégorie se vendent à pertes
```

```python
df_cat_sub_2 = (df.groupby(['Category', 'Sub-Category', 'Month', 'Year'], as_index = False).agg({
    'Amount':'sum',
    'Quantity': 'sum',
    'Profit': 'sum'   
    })
)
df_cat_sub_2['Margin'] = df_cat_sub_2['Profit'] / df_cat_sub_2['Amount']
df_cat_sub_2
```

```python
df.to_csv('e_commerce_cleaned.csv', index = False)
df_cat_sub_2.to_csv('kpi_category_subcategory.csv', index = False)
```
